```python
import pandas as pd

df=pd.read_csv("set-datos.csv")
df
```

```python
def obtener_csv_como_lista_de_diccionarios(nombre_archivo):
  separador = ","
  with open(nombre_archivo, encoding="utf-8") as archivos:
      next(archivos)
      inventario = []

      for linea in archivos:
        linea = linea.rstrip("\n")
        columnas = linea.split(separador)
        nombre = columnas[0]
        cantidad = int(columnas[1])
        precio = float(columnas[2])
        inventario.append({
            "nombre": nombre,
            "cantidad": cantidad,
            "precio": precio
        })
      return inventario

def guardar_inventario(nombre_archivo, inventario):
  with open(nombre_archivo, mode='w', encoding="utf-8") as archivo:
    archivo.write("nombre,cantidad,precio\n")
    for item in inventario:
      linea = f"{item['nombre']},{item['cantidad']},{item['precio']}\n"
      archivo.write(linea)

def mostrar_inventario(inventario):
  for item in inventario:
    print(f"Nombre: {item['nombre']}, Cantidad: {item['cantidad']}, Precio: {item['precio']}")

def realizar_compra(inventario):
  nombre_producto = input("Ingrese el nombre del producto que desea comprar: ")
  cantidad = int(input("Ingrese la cantidad que desea comprar: "))

  for item in inventario:
    if item['nombre'].lower() == nombre_producto.lower():
      if item['cantidad'] >= cantidad:
         item['cantidad'] -= cantidad
         print(f"Ha comprado {cantidad} de {item['nombre']}")
      else:
         print("No hay suficientes existencias")
      break
  else:
      print("Producto no encontrado")


def agregar_productos(inventario):
    nombre = input("Ingrese el nombre del nuevo producto: ")
    cantidad = int(input("Ingrese la cantidad del nuevo producto: "))
    precio = float(input("Ingrese el precio del nuevo producto: "))

    nuevo_producto = {'nombre': nombre, 'cantidad':cantidad, 'precio':precio}
    inventario.append(nuevo_producto)
    print(f"Producto {nombre} agregado al inventario")


def modificar_producto(inventario):
    nombre_producto = input("Ingrese el nombre del producto que desea modificar: ")

    for item in inventario:
        if item['nombre'].lower() == nombre_producto.lower():
            nombre = input(f"Nombre actual: {item['nombre']}. Ingrese el nuevo nombre (o presione Enter para mantener)")
            cantidad = input(f"Cantidad actual: {item['cantidad']}. Ingrese la nueva cantidad (o presione Enter para mantener)")
            precio = input(f"Precio actual {item['precio']}. Ingrese el nuevo precio (o presione Enter para mantener)")

            if nombre:
                item['nombre'] = nombre
            if cantidad:
                item['cantidad'] = int(cantidad)
            if precio:
                item['precio'] = float(precio)

            print(f"Producto {nombre_producto} modificado")
            break
    else:
        print("Producto no encontrado")


def eliminar_producto(inventario):
    nombre_producto = input("Ingrese el nombre del producto que desea eliminar: ")

    for item in inventario:
        if item['nombre'].lower() == nombre_producto.lower():
            inventario.remove(item)
            print(f"Producto {item['nombre']} eliminado del inventario")
            break
    else:
        print("Producto no encontrado")


inventario = obtener_csv_como_lista_de_diccionarios('set-datos.csv')

while True:
    print("\nOpciones:")
    print("1. Mostrar inventario")
    print("2. Realizar una compra")
    print("3. Agregar un producto")
    print("4. Modificar un producto")
    print("5. Eliminar un producto")
    print("6. Guardar y salir")

    opcion = input("Seleccione una opción: ")

    if opcion == '1':
        mostrar_inventario(inventario)
    elif opcion == '2':
        realizar_compra(inventario)
    elif opcion == '3':
        agregar_productos(inventario)
    elif opcion == '4':
        modificar_producto(inventario)
    elif opcion == '5':
        eliminar_producto(inventario)
    elif opcion == '6':
        guardar_inventario('inventario.csv', inventario)
        print("Cambios guardados. Saliendo del programa.")
        break
    else:
        print("Opción no válida. Intente de nuevo.")





```
